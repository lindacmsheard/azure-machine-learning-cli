```python
file = '../components/get_planetary_data/test/example_output_file.tif'
file = '../components/get_planetary_data/test/mlruns/0/aba3bc13379a4bb6aaee2eae1e2cdd63/artifacts/S2A_MSIL2A_20190717T213531_R086_T06VVN_20201005T152616.tif'
file = '../components/get_planetary_data/test/mlruns/0/aba3bc13379a4bb6aaee2eae1e2cdd63/artifacts/S2B_MSIL2A_20190629T212529_R043_T06VVN_20201006T080531.tif'
window = (0,0,2000,2000)
```

```python
#https://stackoverflow.com/questions/67390765/display-specific-part-of-tiff-image-using-rasterio-without-having-to-load-the-en
import rasterio
from rasterio.enums import ColorInterp
from rasterio.plot import show
from rasterio.windows import Window

with rasterio.open(file) as img:
    window = Window(*window)

    # Lookup table for the color space in the source file
    source_colorinterp = dict(zip(img.colorinterp, img.indexes))

    # Read the image in the proper order so the numpy array will have the colors in the
    # order expected by matplotlib (RGB)
    rgb_indexes = [
        source_colorinterp[ci]
        for ci in (ColorInterp.red, ColorInterp.green, ColorInterp.blue)
    ]
    data = img.read(rgb_indexes, window=window)

    # Also pass in the affine transform corresponding to the window in order to
    # display the correct coordinates and possibly orientation
    show(data, transform=img.window_transform(window))
```
